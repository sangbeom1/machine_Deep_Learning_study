```python
# Series

# ndarray에 추가적인 구조를 덧붙여놓은 자료구조라고 생각하시면되요!
# 추가적인 인덱스를 사용할 수 있어요!

# Series의 생성
import numpy as np
import pandas as pd

s = pd.Series([-1, 3, 5, 10])
print(s)

print(s.values) # [-1  3  5 10]
print(s.index) # RangeIndex(start=0, stop=4, step=1)
```

```python
# ndarray는 당연히 숫자 index만 사용할 수 있어요!
# Series는 숫자인덱스와 지정인덱스를 사용할 수 있어요!

s = pd.Series([-1, 5, 10, 40],
              dtype=np.int32,
              index=['c', 'a', 'b', 'k'])
print(s)
# index를 지정인덱스로 설정한 경우 숫자인덱스는 사용할 수 있나요?
print(s[1])  # 5   => 숫자 인덱스도 사용할 수 있어요!
print(s['b']) # 10  => 지정 인덱스로 indexing을 할 수 있어요!
```

```python
s = pd.Series([-1, 5, 10, 40],
              dtype=np.int32,
              index=[0, 2, 100, 6])
print(s)
print(s[100])  # 10
```

```python
# 이번에는 지정인덱스에 중복이 있을 수 있나요?
# 네..지정인덱스는 중복이 가능해요!
s = pd.Series([-1, 5, 10, 40],
              dtype=np.int32,
              index=['c', 'a', 'c', 'k'])
print(s)
print(s['c'])  # 2개를 찾고 그 결과는 Series
```

```python
# 이런 Series도 indexing, slicing, boolean indexing, fancy indexing
# 할 수 있나요?
s = pd.Series([-1, 5, 10, 40],
              dtype=np.int32,
              index=['c', 'a', 'b', 'k'])

print(s)
# print(s[0:3])
# c    -1
# a     5
# b    10

# 지정인덱스로 slicing할 경우 약간 다르게 동작해요!
# print(s['c':'b'])
# c    -1
# a     5
# b    10

# boolean indexing
print(s[s % 2 ==0])

# fancy indexing
print(s[[0,2]])

# 집계함수도 같나요??? => 당연히 똑같아요!
print(s.sum())  # 54
```

```python
# 간단하게 Series의 연산에 대해
# 예제를 통해 알아보아요!
# Series를 만들건데.. index값을 날짜로 할꺼예요!

import numpy as np
import pandas as pd
from datetime import date, datetime, timedelta

start_day = datetime(2020,1,1)
print(start_day)  # 2020-01-01 00:00:00

# python의 list를 만드는 방법중에.
# list comprehension이 있어요!
factory_a = pd.Series([int(x) for x in np.random.normal(50,2,(10,))],
                      index=[start_day + timedelta(days=x) for x in range(10)])
print(factory_a)

start_day = datetime(2020,1,5)
factory_b = pd.Series([int(x) for x in np.random.normal(70,8,(10,))],
                      index=[start_day + timedelta(days=x) for x in range(10)])
print(factory_b)

print(factory_a + factory_b)
```

```python
# Pandas의 DataFrame을 생성해 보아요!
# 가장 간단한 방법은 Dictionary를 이용하는 거예요!

data = {'name': ['홍길동', '신사임당', '강감찬', '이순신'],
        'year': [2020, 2021, 2022, 2020],
        'point': [3.5, 2.0, 1.3, 4.3]}

df = pd.DataFrame(data)
# Jupyter notebook에서 DataFrame을 출력할때 display를 이용
display(df)
print(df.ndim)  # 2 (DataFrame은 2차원이예요! Series는 1차원이예요!)
print(df.shape) # (4,3)
print(df.index) # RangeIndex(start=0, stop=4, step=1)
print(df.values)
# [['홍길동' 2020 3.5]
#  ['신사임당' 2021 2.0]
#  ['강감찬' 2022 1.3]
#  ['이순신' 2020 4.3]]
print(df.columns)
# Index(['name', 'year', 'point'], dtype='object')
```

```python
# 이런 DataFrame을 어떻게 제어할 수 있나요??
# 이 방법들에 대해서 천천히 알아보아요!

# DataFrame의 index와 column에 이름을 지정하는 방법부터 알아보아요!

data = {'names' : ['아이유', '김연아', '홍길동', '강감찬', '이순신'],
        'year': [2020, 2021, 2019, 2022, 2021],
        'point': [3.5, 2.1, 3.1, 1.1, 2.7]}

df = pd.DataFrame(data)
display(df)

# 기존 DataFrame의 형태를 변경해서 새로운 DataFrame을 만들어요!
# inplace라는 속성은 만약 True이면 원본 DataFrame을 변경하고 리턴하지
# 않아요. 만약 False라면 원본을 변경하지 않고 변경된 복사본을 리턴해요!
new_df = df.rename(columns={'year':'입학연도',
                            'point': '학점'},
                   index={0: 'one'},
                   inplace=False)
display(new_df)
```

```python
# 위에서 살펴본 것처럼 DataFrame의 row indexing은
# 너무 복잡해요! 어떤건되고 어떤건 안되요!!
# 사용하기가 껄끄러워요!
# 그래서 일반적으로 DataFrame의 indexing(row)은
# loc를 사용해요!
data = {'이름' : ['아이유', '김연아', '홍길동', '강감찬', '이순신'],
        '입학연도': [2020, 2021, 2019, 2022, 2021],
        '학점': [3.5, 2.1, 3.1, 1.1, 2.7],
        '학과': ['컴퓨터', '철학', '경제', '기계', '물리']}

df = pd.DataFrame(data,
                  columns=['입학연도','이름','학과','학점'],
                  index=['one','two','three','four','five'])
display(df)
# loc를 이용해서 DataFrame의 row indexing을 할 수 있어요!
# 그런데 loc를 이용하려면 반드시 지정 index를 사용해야 해요!
# 만약 기본으로 제공되는 숫자 index를 이용하려면
# loc가 아니라 iloc를 이용해야 해요!

# 하나의 row만 가져올 경우(단일 row만 추출)
# print(df.loc['two']) # Series로 추출되요!

# 슬라이싱도 되요
# display(df.loc['two':'four']) # DataFrame으로 추출되요!

# 슬라이싱할때 지정인덱스와 숫자인덱스 혼합사용은 안되요!
# display(df.loc['two':-1]) # 에러!!

# Fancy indexing도 되요!
# display(df.loc[['two', 'five']]) # DataFrame으로 추출되요!

# 숫자 index를 사용할경우 iloc를 이용해야 해요!
display(df.iloc[[1,3]]) # DataFrame으로 추출되요!
```

```python
# 5. sort()라는 함수는 당연히 정렬하는 함수예요!
#    데이터프레임(Series)의 내용을 내가 원하는 조건으로 정렬시키요!

import numpy as np
import pandas as pd

# 난수(랜덤값)은 우리가 생각하는 랜덤값과 약간 달라요!
# 세상살때의 랜덤값은 진짜 랜덤값이예요!
# 컴퓨터에서 발생시키는 랜덤값은 진짜 랜덤값일까?
# 사실 컴퓨터의 랜덤값은 특정 알고리즘을 기반으로 도출된 값이예요!
# 이 알고리즘의 기반이 되는 값이 있는데 이 값을 seed라고 해요!
# 이 seed값을 고정시키면 발생되는 난수값이 고정되요!
np.random.seed(3)

df = pd.DataFrame(np.random.randint(0,10,(6,4)))

df.columns = ['A', 'B', 'C', 'D']
df.index = pd.date_range('20200101', periods=6)
# display(df)
# 이렇게 DataFrame을 만들었어요!!
# 이제 정렬을 좀 알아볼꺼예요!

# 만들어진 DataFrame의 column의 순서와 row의 순서를
# 섞어볼꺼예요!
random_index = np.random.permutation(df.index)
# print(random_index)

df2 = df.reindex(index=random_index,
                 columns=['B','A','D','C'])
display(df2)

# 일단 이렇게 DataFrame의 내용을 마구마구 섞었어요!
# 정렬해 보아요! 정렬의 기본은 당연히 오름차순이예요!
# 오름차순은 작은게 위로, 작은게 앞으로 오는게 오름차순이죠.
# sort_index(), sort_values()
# display(df2.sort_index(axis=1))
# axis=0 이면 행을 기준으로 정렬
# axis=1 이면 열(column)을 기준으로 정렬해요!
df2.sort_values(by='A')

# index, 혹은 값을 이용해서 정렬할 수 있어요!
```
