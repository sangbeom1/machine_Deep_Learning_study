```python
# Pandas의 나머지 기능들에 대해서 알아보아요!
# Pandas가 가지는 집계함수가 있어요!
# Numpy의 집계함수와 비슷합니다.!!
# sum(), mean(), max(), count(), ...

# DataFrame을 가지고 알아보아요!
import numpy as np
import pandas as pd

data = [[2, np.nan],
        [7, -3],
        [np.nan, np.nan],
        [1,2]]

df = pd.DataFrame(data,
                  columns=['one','two'],
                  index=['a', 'b', 'c', 'd'])
display(df)

# 집계함수를 이용해 보아요!
# 가장 대표적인 함수는 sum(), mean()

# print(df.sum())
# one    10.0
# two    -1.0
# dtype: float64
# 우리 예상과는 다르게 몽땅더하는게 아니라..
# axis=0 으로 간주되어서 실행됩니다.
# numpy(ndarray)는 집계함수 이용시 만약 axis를 명시하지 않으면
# axis=None으로 간주하는데
# pandas의 DataFrame은 만약 axis를 명시하지 않으면
# axis=0으로 간주해요!(결과는 Series로 리턴되요!)

# print(df.mean(axis=0, skipna=True)) # NaN은 아예 무시되요!

# 만약 skipna를 False로 설정하면 어떻게 되나요?
# print(df.mean(axis=0, skipna=False))
# NaN은 실수로 간주하고 연산이 가능, 단 값을 알수 없어요!라는 의미이기
# 때문에 NaN과 연산하면 무조건 결과는 NaN으로 나와요!

# 이렇게 우리가 만든 DataFrame이 있는데
# 이 DataFrame안의 데이터를 사용하기위해서(데이터 분석, 머신러닝의 입력)
# 이 두 경우 모두
# 일반적으로 NaN이 있으면 연산이 안되요!
# 이런 NaN을 우리가 결치값(결치, missing value)라고 부르는데
# 당연히 이런 값을 처리해줘야 해요!
# 처리하는 방법은 크게 2가지 있어요!
# 1. 가장 좋은 방법은 NaN을 삭제하는 거예요!
#    그런데 문제가 있어요!
#    우리가 일반적으로 데이터분석이나 머신러닝 학습 데이터로 사용할 데이터는
#    이 데이터는 많은 많을수록 좋아요!
#    많을수록 좋다고 하는데 일반적으로 많다는 기준은 그럼 얼마인가요?
#    사실 문제(domain)마다 달라요. 하지만 약 10만건을 기준으로 잡아요!
#    10만개를 기준으로 그거보다 적으면 데이터가 작아요!
#    10만개를 기준으로 그거보다 많은면 데이터가 많이 있어요!
#    현실적으로는 데이터를 그렇게 많이 확보하기가 쉽지 않아요!
#    그렇다는 얘기는 가뜩이나 데이터가 적다는 말이예요!
#    없는 와중에 NaN이 포함되어있는 행을 모두 지우면 당연하게 데이터량이 감소
#    결론적으로 데이터의 전체 양이 많은 경우 삭제하는게 가장 좋아요!
#    하지만 데이터량이 적은경우에는 삭제하지 않는게 좋아요!
# 2. 그럼 다른 방법이 있나요? imputation을 사용해요!
#    NaN으로 되어 있는 놈을 값을 변화시켜서 사용해요!
#    그럼 어떻게 변환시킬까요?
#    2-1. 집계함수를 이용(기본통계기법)해서 값을 대체 -> 평균
#    2-2. 머신러닝 기법을 이용해서 값을 대체 -> 추론된 값을 이용

# pandas로 이런 결측치를 처리해야 해요!
# 결측치 삭제는 아래처럼 작업해요!
# new_df = df.dropna(how='any')
# display(new_df)

# 결측치를 이번에는 다른 값으로 대체
one_avg = df['one'].mean()  # 3.3333
two_min = df['two'].min() # -3

df['one'] = df['one'].fillna(value=one_avg)
df['two'] = df['two'].fillna(value=two_min)
display(df)
```

```python
# 이번에는 DataFrame을 합치는 방법에 대해서 알아보아요!
# 마치 데이터베이스 안의 테이블간의 연동을 떠올리시면 되요! (join)

# 많은 경우에 데이터가 여러개의 DataFrame으로 나뉘어져 있는 경우가 있어요!
# 많은 DataFrmae이 있을 때 이 DataFrame을 그냥 냅다 연결시키는 방법이 있어요!
# 엑셀로 간단히 알아보았어요!

# 코드로 살펴보아요!

df1 = pd.DataFrame({'a': ['a0', 'a1', 'a2', 'a3'],
                    'b': [1, 2, 3, 4],
                    'c': ['c0', 'c1', 'c2', 'c3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'b': ['b0', 'b1', 'b2', 'b3'],
                    'c': ['c0', 'c1', 'c2', 'c3'],
                    'd': ['d0', 'd1', 'd2', 'd3'],
                    'e': ['e0', 'e1', 'e2', 'e3']},
                   index=[2, 3, 4, 5])
display(df1, df2)

result1 = pd.concat([df1, df2], axis=0)
display(result1)

# 이렇게 concat()을 이용하면 row index는 그냥 덧붙이게되서
# index가 중복될 수 있어요! 그래서 다음과 같이 하면
# index를 새롭게 지정할 수 있어요!
result2 = pd.concat([df1, df2], axis=0, ignore_index=True)
display(result2)
```

```python
# 이렇게 DataFrame을 아무생각없이 연결할 수 있어요!
# 행방향, 열방향으로 연결할 수 있습니다.
# 단, 연결할때 결과가 어떻게 나오는지는 주의해야 해요!

# 이번에는 DataFrame을 연결하는 또 다른 방법에 대해서 알아보아요!
# join에 대해서 알아보아요!

data1 = { "학번" : [1,2,3,4],
          "이름" : ["이지안","박동훈","이순신","강감찬"],
          "학년" : [2,4,1,3]}

data2 = { "학번" : [1,2,4,5],
          "학과" : ["CS","MATH","MATH","CS"],
          "학점" : [3.4,2.9,4.5,1.2]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

display(df1, df2)

# 두개의 DataFrame을 결합(merge)해보아요!
# 먼저 inner join에 대해서 알아보아요!
result = pd.merge(df1,
                  df2,
                  on='학번',
                  how='inner')
display(result)
```

```python
data1 = { "학번" : [1,2,3,4],
          "이름" : ["이지안","박동훈","이순신","강감찬"],
          "학년" : [2,4,1,3]}

data2 = { "학번" : [1,2,4,5],
          "학과" : ["CS","MATH","MATH","CS"],
          "학점" : [3.4,2.9,4.5,1.2]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

display(df1, df2)

# 이번에는 outer join에 대해서 알아보아요!
# outer join은 inner join에 포함되지 않는 행들을 결과에 포함시키는
# 방법이예요..
# 그래서 outer join은 크게 3가지가 존재해요!
# left outer join, right outer join, full outer join

result1 = pd.merge(df1, df2, on='학번', how='left')
# 일단 inner join을 하고...왼쪽(df1)에서 남은 행은 결과에 추가
# display(result1)

result1 = pd.merge(df1, df2, on='학번', how='right')
# 일단 inner join을 하고...오른쪽(df2)에서 남은 행은 결과에 추가
# display(result1)

# 마지막으로 full outer join은 how를 outer로 명시하면 되요!
result1 = pd.merge(df1, df2, on='학번', how='outer')
```

```python
# 이번에는 titanic data set을 이용해서
# grouping을 알아보아요!

titanic = sns.load_dataset('titanic')
# display(titanic.head())

df = titanic.loc[:, ['age', 'sex', 'class', 'fare', 'survived']]
# display(df.head())  # 총 891개의 데이터가 존재해요!

# 그룹으로 묶어보아요
# class를 기준으로 그룹을 묶어보아요!
grouped = df.groupby(['class'])  # 3개의 그룹으로 행들이 묶이게 되요!
# print(grouped)  # 이렇게는 그룹에 대한 내용을 확인할 수 없어요!

# 현재 grouped는 3그룹이예요.. 당연히 loop는 3번 돌겠죠.
# for key, group in grouped:
#     print(key)
#     display(group.head(3))

# 아하..이렇게 각 group으로 되어있는 DataFrame을 우리가 얻을 수 있어요!
# 일반적으로 이렇게 group으로 묶은뒤에 우리는 집계함수를 이용해서
# 통계치를 얻게되요!

# 각 그룹의 평균을 알아보아요~!
display(grouped.mean())

# 특정 그룹에 대한 값을 알고 싶어요!
# group3 = grouped.get_group('Third')
# display(group3.head())
```
